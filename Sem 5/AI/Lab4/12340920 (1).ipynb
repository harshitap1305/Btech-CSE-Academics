```python
import math

# -----------------------------
# Helper functions
# -----------------------------
def is_terminal(state):
    return all(h == 0 for h in state)

def get_moves(state):
    moves = []
    for i, heap in enumerate(state):
        for remove in range(1, heap + 1):  # always remove fewer stones first
            new_state = list(state)
            new_state[i] -= remove
            moves.append((state, tuple(new_state)))
    return moves

# -----------------------------
# Alpha-Beta with Memoization
# -----------------------------
nodes_explored = 0
nodes_pruned = 0
visited = {}

def alphabeta(state, alpha, beta, maximizing_player=True):
    global nodes_explored, nodes_pruned, visited

    key = (state, maximizing_player)
    if key in visited:
        return visited[key]

    nodes_explored += 1

    if is_terminal(state):
        value = -1 if maximizing_player else 1
        visited[key] = value
        return value

    if maximizing_player:
        print(f"MAX explores {state}")
        value = -math.inf
        for move in get_moves(state):
            print(f" Considering move: {move[0]} → {move[1]}")
            child_val = alphabeta(move[1], alpha, beta, False)
            value = max(value, child_val)
            alpha = max(alpha, value)
            if alpha >= beta:
                print(f" [Pruned branch at {move[1]} because alpha >= beta]")
                nodes_pruned += 1
                break
        visited[key] = value
        return value
    else:
        print(f"MIN explores {state}")
        value = math.inf
        for move in get_moves(state):
            print(f" Considering move: {move[0]} → {move[1]}")
            child_val = alphabeta(move[1], alpha, beta, True)
            value = min(value, child_val)
            beta = min(beta, value)
            if beta <= alpha:
                print(f" [Pruned branch at {move[1]} because beta <= alpha]")
                nodes_pruned += 1
                break
        visited[key] = value
        return value

# -----------------------------
# Run the algorithm
# -----------------------------
def run_game(initial_state):
    global nodes_explored, nodes_pruned, visited
    nodes_explored = 0
    nodes_pruned = 0
    visited = {}

    print(f"Initial State: {initial_state}")

    best_value = -math.inf
    best_move = None
    alpha, beta = -math.inf, math.inf

    for move in get_moves(initial_state):
        print(f" Considering move: {move[0]} → {move[1]}")
        value = alphabeta(move[1], alpha, beta, False)
        if value > best_value:
            best_value = value
            best_move = move
        alpha = max(alpha, best_value)

    print(f"\nBest Move for MAX: {best_move[0]} → {best_move[1]}")
    if best_value == 1:
        print("Outcome: Winning position")
    else:
        print("Outcome: Losing position")
    print(f"Nodes Explored: {nodes_explored}")
    print(f"Nodes Pruned: {nodes_pruned}")

# -----------------------------
# Example Run
# -----------------------------
initial_state = (3, 4, 5)
run_game(initial_state)
```
