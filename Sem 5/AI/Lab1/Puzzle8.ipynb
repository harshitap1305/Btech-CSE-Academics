```python
import heapq
```

```python
class Puzzle:
    def __init__(self):
        self.n = 3
        self.goal = ((1, 2, 3), (4, 5, 6), (7, 8, 0))
        self.moves = [(-1, 0), (1, 0), (0, -1), (0, 1)]

    def manhattan(self, state):
        return sum(
            abs(i - (val - 1) // self.n) + abs(j - (val - 1) % self.n)
            for i, row in enumerate(state)
            for j, val in enumerate(row)
            if val != 0
        )
    
   #this converts a 2D list → immutable tuple of tuples for use in sets.
    def serialize(self, state):
        return tuple(tuple(row) for row in state)

    def print_path(self, node):
        path = []
        while node:
            path.append(node[1])
            node = node[5]  # parent
        for state in reversed(path):
            for row in state:
                print(*(x if x else "_" for x in row))
            print()

    def solve(self, start):
        start_state = self.serialize(start)
        start_h = self.manhattan(start)
        blank_pos = next((i, j) for i in range(self.n) for j in range(self.n) if start[i][j] == 0)

        # heap elements: (f, state, g, blank_pos, h, parent)
        heap = [(start_h, start_state, 0, blank_pos, start_h, None)]
        visited = set()

        while heap:
            f, state, g, (br, bc), h, parent = heapq.heappop(heap)

            if state == self.goal:
                print("Solution found:\n")
                self.print_path((f, state, g, (br, bc), h, parent))
                return

            if state in visited:
                continue
            visited.add(state)

            for dr, dc in self.moves:
                nr, nc = br + dr, bc + dc
                if 0 <= nr < self.n and 0 <= nc < self.n:
                    new_state = [list(row) for row in state]
                    new_state[br][bc], new_state[nr][nc] = new_state[nr][nc], new_state[br][bc]
                    new_state = self.serialize(new_state)

                    if new_state in visited:
                        continue

                    h2 = self.manhattan(new_state)
                    heapq.heappush(heap, (g + 1 + h2, new_state, g + 1, (nr, nc), h2, (f, state, g, (br, bc), h, parent)))

        print("No solution found.\n")
```

```python
start = [
    [1, 2, 3],
    [4, 0, 6],
    [7, 5, 8]
]

p = Puzzle()
p.solve(start)
```
