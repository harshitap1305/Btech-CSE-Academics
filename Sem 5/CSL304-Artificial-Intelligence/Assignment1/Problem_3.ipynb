```python
import math

def winner(b, p):
    for i in range(3):
        if b[i][0] == p and b[i][1] == p and b[i][2] == p:
            return True
        if b[0][i] == p and b[1][i] == p and b[2][i] == p:
            return True
    if b[0][0] == p and b[1][1] == p and b[2][2] == p:
        return True
    if b[0][2] == p and b[1][1] == p and b[2][0] == p:
        return True
    return False

def full(b):
    for i in range(3):
        for j in range(3):
            if b[i][j] == " ":
                return False
    return True

def evaluate(b):
    X1 = X2 = X3 = O1 = O2 = O3 = 0
    lines = []

    for i in range(3):
        lines.append(b[i])
        lines.append([b[0][i], b[1][i], b[2][i]])
    lines.append([b[0][0], b[1][1], b[2][2]])
    lines.append([b[0][2], b[1][1], b[2][0]])

    for line in lines:
        if line.count("X") > 0 and line.count("O") == 0:
            if line.count("X") == 1: X1 += 1
            elif line.count("X") == 2: X2 += 1
            elif line.count("X") == 3: X3 += 1
        if line.count("O") > 0 and line.count("X") == 0:
            if line.count("O") == 1: O1 += 1
            elif line.count("O") == 2: O2 += 1
            elif line.count("O") == 3: O3 += 1

    return 8*X3 + 3*X2 + X1 - (8*O3 + 3*O2 + O1)

def minimax(b, isMax, alpha, beta, tree, move="Root"):
    if winner(b, "X") or winner(b, "O") or full(b):
        val = evaluate(b)
        tree.append((move, val))
        return val

    if isMax:
        best = -math.inf
        for i in range(3):
            for j in range(3):
                if b[i][j] == " ":
                    b[i][j] = "X"
                    child = []
                    val = minimax(b, False, alpha, beta, child, f"X({i},{j})")
                    tree.append((move, None, child))
                    b[i][j] = " "
                    best = max(best, val)
                    alpha = max(alpha, best)
                    if beta <= alpha:
                        break
        return best
    else:
        best = math.inf
        for i in range(3):
            for j in range(3):
                if b[i][j] == " ":
                    b[i][j] = "O"
                    child = []
                    val = minimax(b, True, alpha, beta, child, f"O({i},{j})")
                    tree.append((move, None, child))
                    b[i][j] = " "
                    best = min(best, val)
                    beta = min(beta, best)
                    if beta <= alpha:
                        break
        return best

def print_tree(tree, indent=0):
    for node in tree:
        if len(node) == 2:  # terminal
            print(" " * indent + f"{node[0]} → {node[1]}")
        else:
            print(" " * indent + f"{node[0]}")
            print_tree(node[2], indent+4)

board = [
    [' ', 'X', ' '],
    ['O', ' ', ' '],
    ['X', ' ', 'O']
]

best_val = -math.inf
best_move = None
main_tree = []

for i in range(3):
    for j in range(3):
        if board[i][j] == " ":
            board[i][j] = "X"
            subtree = []
            move_val = minimax(board, False, -math.inf, math.inf, subtree, f"X({i},{j})")
            board[i][j] = " "
            if move_val > best_val:
                best_val = move_val
                best_move = (i, j)
            main_tree.append(("Root", None, subtree))

print(f"Best Move for X: {best_move} with value: {best_val}\n")
print("Game Tree:")
print_tree(main_tree)
```
