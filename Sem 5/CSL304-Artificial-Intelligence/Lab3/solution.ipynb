```python
import random
import math
```

N: number of queens 

```python
N=8
```

```python
# function for user input of initial state
def user_input_state():
    print(f"Enter {N} numbers between 1 and {N} (space-separated),")
    print("Each number represents the row of the queen in that column: ")
    try: 
        state=list(map(int,input("Initial state: ").split()))
        if len(state)!=N or any (x<1 or x>N for x in state):
            raise ValueError
        return state
    except:
        print("Invalid input. Using random state instead")
        return [random.randint(1,N) for _ in range(N)]
```

```python
# function for user inout of parameters: temperature and cooling rate
def user_input_params():
    try:
        temp=float(input("enter start temp (eg 100): "))
        cooling=float(input("enter cooling rate b/w 0 and 1: "))
        if not(0<cooling<1):
            raise ValueError
        return temp,cooling
    except:
        print("Invalid input, using default (temp=100,cooling=0.99) ")
        return 100,0.99
```

```python
#function to count no. of conflicts
def conflicts(state):
    count=0
    for i in range (N):
        for j in range(i+1,N):
            if state[i]==state[j]:
                count+=1
            if abs(state[i]-state[j]) == abs(i-j):
                count+=1
    return count
```

```python
#function to generate neighbor by moving one queen to a diff row
def neighbor(state):
    new_state=state[:]
    col=random.randint(0,N-1)
    row=random.randint(1,N)
    while row == state[col]:
        row=random.randint(1,N)
    new_state[col]=row
    return new_state
```

```python
#simulated annealing algorithm for N-Queens
def simulated_annealing (initial_state, max_iterations=100000, start_temp=100,cooling_rate=0.99):
    current=initial_state[:]
    current_conflicts=conflicts(current)
    temperature=start_temp
    for step in range(max_iterations):
        if current_conflicts == 0:
            return current, step
        new =neighbor(current)
        new_conflicts=conflicts(new)
        delta=new_conflicts-current_conflicts

        if delta<0 or random.random() < math.exp(-delta/temperature):
            current,current_conflicts =new,new_conflicts

        temperature*=cooling_rate
        if temperature < 1e-5:
            break
    return current,max_iterations
```

```python
# helper to print chess board

def print_board(state):
    for row in range(1,N+1):
        line=""
        for col in range(N):
            if state[col]==row:
                line+='Q'
            else:
                line+='.'
        print(line)
    print()
```

```python
#main func
if __name__ == "__main__":
    initial_state=user_input_state()
    start_temp,cooling_rate=user_input_params()
    print("\nStarting State:", initial_state, "Conflicts:", conflicts(initial_state))
    print(f"Parameters: Temperature={start_temp}, Cooling Rate={cooling_rate}\n")

    solution, steps = simulated_annealing(initial_state, start_temp=start_temp, cooling_rate=cooling_rate)

    print("Final Solution:", solution)
    print("Conflicts:", conflicts(solution))
    print("Steps taken:", steps)
    print("\nBoard Representation:")
    print_board(solution)
```
